```python
import logging
import gensim
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
id2word = gensim.corpora.Dictionary.load_from_text('wiki_wordids.txt.bz2')
mm = gensim.corpora.MmCorpus('wiki_tfidf.mm')

lda = gensim.models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=100, update_every=1, passes=1)
```

```python
lda.save('lda.model')
```

```python
lda_copy = gensim.models.LdaModel.load('lda.model')
```
